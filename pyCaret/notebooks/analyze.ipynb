# Analyze

Esploriamo le funzioni di analisi e explainability messe a disposizione da PyCaret. 

## `plot_model`
Una funzione che analizza le performance di un modello allenato su di un dataset hold-out. 

```python
# load dataset
from pycaret.datasets import get_data
diabetes = get_data('diabetes')

# init setup
from pycaret.classification import *
clf1 = setup(data = diabetes, target = 'Class variable')

# creating a model
lr = create_model('lr')

# plot model
plot_model(lr, plot = 'auc')
```

```python
# load dataset
from pycaret.datasets import get_data
diabetes = get_data('diabetes')

# init setup
from pycaret.classification import *
clf1 = setup(data = diabetes, target = 'Class variable')

# creating a model
lr = create_model('lr')

# plot model
plot_model(lr, plot = 'auc', scale = 3)
```

```python
# plot model
plot_model(lr, plot = 'auc', save = True)
```

## Diversi tipi di plot

Possiamo creare la confusion matrix come segue. In generale, PyCaret utilizza Yellowbrick per la visualizzazione, e quindi i parametri che possono essere presi da Yellowbrick possono essere passati al metodo `plot_model` tramite utilizzo del parametro `plot_kwardgs`.

```python
# plot model
plot_model(lr, plot = 'confusion_matrix')
```

```python
# plot con percentuali
plot_model(lr, plot = 'confusion_matrix', plot_kwargs = {'percent' : True})
```
